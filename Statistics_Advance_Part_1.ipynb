1. What is a random variable in probability theory?
A random variable is a variable that takes on numerical values based on the outcomes of a random phenomenon. It maps outcomes of a probability experiment to numbers.

2. What are the types of random variables?
There are two main types:
Discrete random variables – take on a countable number of values (e.g., number of heads in coin flips).
Continuous random variables – take on values from a continuous range (e.g., time, weight).

3. What is the difference between discrete and continuous distributions?
Discrete distributions describe random variables with specific, countable outcomes (e.g., 0, 1, 2…).
Continuous distributions describe random variables with an uncountable range of possible values, often real numbers in an interval.

4. What are probability distribution functions (PDF)?
For discrete variables, this is called the probability mass function (PMF), giving the probability of each specific value.
For continuous variables, the probability density function (PDF) describes the relative likelihood of a variable taking a value in an interval; probabilities are found via area under the curve.

5. How do cumulative distribution functions (CDF) differ from probability distribution functions (PDF)?
A PDF gives the instantaneous likelihood (for continuous) or exact probabilities (for discrete).
A CDF gives the cumulative probability that a random variable is less than or equal to a given value.

6. What is a discrete uniform distribution?
A discrete uniform distribution assigns equal probability to each outcome in a finite set. Example: Rolling a fair 6-sided die (each face has 1/6 chance).

9. What is the Poisson distribution and where is it applied?
A Poisson distribution models the number of events in a fixed interval of time or space when events happen at a constant average rate and independently.
Used for modeling rare events (e.g., phone calls per hour at a call center).

10. What is a continuous uniform distribution?
A continuous uniform distribution is one where all values in a continuous interval
[a,b] are equally likely.

11. What are the characteristics of a normal distribution?
A normal distribution is a symmetric, bell-shaped curve where most values cluster around the mean. For example, adult human heights often follow a normal distribution—if the average height is 170 cm, most people will be around that height, with fewer extremely short or tall individuals.

12. What is the standard normal distribution, and why is it important?
The standard normal distribution has a mean of 0 and standard deviation of 1. It's important because any normal distribution can be converted into it using Z-scores. For instance, a test score of 85 in a class with mean 75 and SD 5 becomes a Z-score of 2 (2 SDs above average), which we can interpret using standard normal tables

13. What is the Central Limit Theorem (CLT), and why is it critical in statistics?
The CLT says that the average of a large number of independent, identically distributed variables will be normally distributed. For example, even if individual incomes are skewed, the average income of 50 people will follow a roughly normal distribution, allowing reliable inferences.

15. What is the application of Z statistics in hypothesis testing?
Z-statistics test whether a sample comes from a population with a certain mean. For example, if we want to test if a new teaching method improves scores (mean 80 vs. old 75), we calculate a Z-score. If it’s far enough from 0 (e.g., beyond ±1.96), we reject the null hypothesis.

17. What are point estimates and interval estimates in statistics?
A point estimate is a single value—like a sample mean of 200 mg—for estimating a population parameter. An interval estimate, such as "200 ± 10 mg", provides a range where the true mean likely lies, adding context and confidence to the estimate.

18. What is the significance of confidence intervals in statistical analysis?
Confidence intervals give a range of plausible values for a population parameter. For example, a 95% CI of [68, 72] for average IQ means we’re 95% confident the true average falls in that range, helping guide decisions with uncertainty in mind.

19.  What is the relationship between a Z-score and a confidence interval?
Confidence intervals use Z-scores to set their range. For instance, a 95% CI for a normal distribution uses ±1.96 Z. If a sample mean is 100 with SE 2, the CI is 100 ± 1.96×2 = [96.08, 103.92].

20. How are Z-scores used to compare different distributions?
Z-scores allow fair comparisons across different scales. For example, scoring 85 on a math test with mean 75, SD 5 (Z = 2), and 72 on an English test with mean 60, SD 6 (Z = 2) shows equal performance relative to peers in both subjects.

21. What are the assumptions for applying the Central Limit Theorem (CLT)?
The CLT assumes that the sample observations are independent, the sample size is sufficiently large (usuallyn≥30), and the population has a finite mean and variance. For example, if you randomly sample the daily sales from 50 different stores, the average will be approximately normally distributed, even if individual store sales aren’t, as long as these assumptions hold.

22.  What is the concept of expected value in a probability distribution?
The expected value is the average outcome you'd expect over many trials of a random event. It's like a weighted average of all possible values. For example, the expected value of a fair 6-sided die roll is 3.5, meaning over many rolls, your average result will approach this number.

23. How does a probability distribution relate to the expected outcome of a random variable?
A probability distribution shows all the possible values a random variable can take and how likely each is, while the expected outcome (or expected value) is the average you’d get based on those probabilities. For example, if a lottery ticket has a 10% chance to win $50 and 90% chance to win nothing, the expected value is $5.
