1. What is hypothesis testing in statistics?
Hypothesis testing is a statistical method used to make decisions or inferences about population parameters based on sample data. It involves testing an assumption (the hypothesis) to determine if there's enough evidence to reject it.

2.  What is the null hypothesis, and how does it differ from the alternative hypothesis?
Null hypothesis (H₀): A statement of no effect or no difference. It's the default or status quo.
Alternative hypothesis (H₁ or Ha): The statement you want to test for; it suggests a real effect or difference.

3. What is the significance level in hypothesis testing, and why is it important?
The significance level (α) is the threshold for rejecting the null hypothesis—commonly set at 0.05 (5%). It represents the probability of making a Type I error (rejecting a true null hypothesis). A lower α means stricter evidence is required to reject H₀.

4. What does a P-value represent in hypothesis testing?
The P-value is the probability of obtaining results at least as extreme as those observed, assuming the null hypothesis is true. It quantifies the strength of evidence against H₀.

5. How do you interpret the P-value in hypothesis testing?
If P ≤ α: Reject the null hypothesis (strong evidence against H₀).
If P > α: Do not reject the null hypothesis (not enough evidence against H₀).

6. What are Type I and Type II errors in hypothesis testing?
Type I Error (α): Rejecting a true null hypothesis (false positive).
Type II Error (β): Failing to reject a false null hypothesis (false negative).

7. What is the difference between a one-tailed and a two-tailed test in hypothesis testing?
One-tailed test: Tests for deviation in one direction (e.g., H₁: μ > 100).
Two-tailed test: Tests for deviation in either direction (e.g., H₁: μ ≠ 100).

8. What is the Z-test, and when is it used in hypothesis testing?
A Z-test is used when:
The sample size is large (n ≥ 30), and
The population standard deviation (σ) is known.
It tests whether a sample mean differs significantly from a population mean.

9. How do you calculate the Z-score, and what does it represent in hypothesis testing?
Z-score formula:
x: Sample mean
μ: Population mean
σ: Population standard deviation
n: Sample size
It represents how many standard deviations the sample mean is from the population mean.

10. What is the T-distribution, and when should it be used instead of the normal distribution?
The T-distribution is used when:
The sample size is small (n < 30), and
The population standard deviation is unknown.
It’s wider and has heavier tails than the normal distribution, accounting for greater uncertainty.

11. Difference between Z-test and T-test
A Z-test is used when the sample size is large and the population standard deviation is known, while a T-test is used when the sample size is small and the population standard deviation is unknown. Z-tests rely on the normal distribution; T-tests use the T-distribution which accounts for more variability in small samples.

12. What is the T-test and its use
A T-test compares the means of one or two groups to see if there’s a statistically significant difference. It’s commonly used when dealing with small samples and when the population standard deviation is not known.

13. Relationship between Z-test and T-test
Both tests assess hypotheses about means, but a T-test is more flexible for small samples. As sample size increases, the T-distribution approaches the Z-distribution, making both tests give similar results in large samples.

14. What is a confidence interval and its use
A confidence interval gives a range within which we expect a population parameter (like a mean) to fall, based on sample data, with a certain level of confidence (e.g., 95%). It reflects the uncertainty in estimating the true value.

15. What is the margin of error and its effect
The margin of error is the amount added to or subtracted from a sample estimate to create a confidence interval. A larger margin of error makes the interval wider, showing greater uncertainty in the estimate.

16. Bayes’ Theorem in statistics
Bayes’ Theorem updates the probability of an event based on new evidence. It's useful in many areas like medical testing and machine learning where prior information is combined with new data to make better predictions.

17.  What is the Chi-square distribution and its use
The Chi-square distribution is used to analyze categorical data and test relationships or variances. It’s commonly applied in goodness-of-fit tests, tests of independence, and in assessing variance differences.

18. What is the Chi-square goodness of fit test
The Chi-square goodness of fit test checks whether observed categorical data matches expected distributions. It helps determine if a model or theoretical distribution fits actual data well.

19. What is the F-distribution and when it’s used
The F-distribution is used to compare variances and is the basis for ANOVA tests. It is right-skewed and used when testing if multiple groups have equal variances or means

20. What is ANOVA and its assumptions
ANOVA (Analysis of Variance) is used to compare the means of three or more groups to see if at least one differs significantly. It assumes data are independent, normally distributed, and have equal variances across groups.

21. What are the different types of ANOVA tests?
There are three main types of ANOVA tests: One-way ANOVA, which compares means across three or more independent groups based on one factor; Two-way ANOVA, which examines the effect of two independent factors and their interaction on a dependent variable; and Repeated Measures ANOVA, used when the same subjects are measured multiple times under different conditions. Each type helps identify whether group differences are statistically significant.

22. What is the F-test, and how does it relate to hypothesis testing?
The F-test is a statistical test used to compare variances between groups and is central to ANOVA. In hypothesis testing, it evaluates whether the variability between group means is significantly greater than the variability within groups. A high F-value indicates that the null hypothesis (that all group means are equal) is unlikely, suggesting at least one group mean is different.

```python
# 14. Calculate Z-Score and Compare to Standard Normal
def calculate_z_score(x, mean, std):
    z = (x - mean) / std
    print(f"Z-score: {z}")

    # Probability of observing a value <= x
    prob = norm.cdf(z)
    print(f"Probability of x <= {x}: {prob}")

    return z

# Example
calculate_z_score(x=72, mean=70, std=5)
```

```python
# 15. Hypothesis Testing Using Z-Statistics
from scipy.stats import norm

def z_test(sample_mean, population_mean, std_dev, sample_size, alpha=0.05):
    # Compute z-statistic
    z = (sample_mean - population_mean) / (std_dev / np.sqrt(sample_size))
    p_value = 1 - norm.cdf(abs(z))  # One-tailed test (for two-tailed, multiply by 2)

    print(f"Z-statistic: {z}")
    print(f"P-value: {p_value}")

    # Decision
    if p_value < alpha:
        print("Reject the null hypothesis (significant)")
    else:
        print("Fail to reject the null hypothesis (not significant)")

# Example:
# Sample mean = 52, population mean = 50, std dev = 5, sample size = 30
z_test(sample_mean=52, population_mean=50, std_dev=5, sample_size=30)
```

```python
# part 2
# 1 Perform a Z-Test for Comparing a Sample Mean to a Known Population Mean
import numpy as np
import scipy.stats as stats

# Parameters
sample_data = np.random.normal(loc=50, scale=10, size=100)  # Sample data from a normal distribution
population_mean = 52  # Known population mean
sample_mean = np.mean(sample_data)
sample_std = np.std(sample_data, ddof=1)  # Sample standard deviation
sample_size = len(sample_data)

# Z-test: (sample mean - population mean) / (sample standard deviation / sqrt(sample size))
z_score = (sample_mean - population_mean) / (sample_std / np.sqrt(sample_size))

# Calculate the p-value (two-tailed test)
p_value = 2 * (1 - stats.norm.cdf(abs(z_score)))  # Multiply by 2 for two-tailed test

# Output
print(f"Z-score: {z_score:.2f}")
print(f"P-value: {p_value:.4f}")

# Interpretation
alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis: The sample mean is significantly different from the population mean.")
else:
    print("Fail to reject the null hypothesis: No significant difference between the sample mean and population mean.")
```

```python
# 4 Perform a Two-Tailed Z-Test and Visualize the Decision Region
import matplotlib.pyplot as plt

# Parameters
population_mean = 50
sample_data = np.random.normal(loc=52, scale=10, size=100)  # Sample data
sample_mean = np.mean(sample_data)
sample_std = np.std(sample_data, ddof=1)
sample_size = len(sample_data)

# Z-score calculation
z_score = (sample_mean - population_mean) / (sample_std / np.sqrt(sample_size))

# Calculate the p-value (two-tailed test)
p_value = 2 * (1 - stats.norm.cdf(abs(z_score)))

# Plot the Z-distribution with decision regions
x = np.linspace(-4, 4, 1000)
y = stats.norm.pdf(x)

plt.plot(x, y, label="Standard Normal Distribution", color="blue")
plt.fill_between(x, y, where=(x <= -1.96), color="red", alpha=0.3)
plt.fill_between(x, y, where=(x >= 1.96), color="red", alpha=0.3)

plt.axvline(x=z_score, color='green', linestyle='--', label=f"Z-score = {z_score:.2f}")
plt.title("Two-Tailed Z-Test - Decision Region")
plt.xlabel("Z-score")
plt.ylabel("Density")
plt.legend()
plt.grid(True)
plt.show()

# Interpretation
if p_value < 0.05:
    print("Reject the null hypothesis: The sample mean is significantly different.")
else:
    print("Fail to reject the null hypothesis: No significant difference.")
```

```python
# 5 Calculate and Visualize Type 1 and Type 2 Errors During Hypothesis Testing
def visualize_type_errors(sample_mean, population_mean, sample_size, alpha=0.05):
    sample_std = 10  # Assume standard deviation
    z_critical = stats.norm.ppf(1 - alpha / 2)  # Z critical value for two-tailed test

    # Type 1 Error (Alpha): False positive (rejecting the true null hypothesis)
    x1 = np.linspace(population_mean - 4 * sample_std, population_mean + 4 * sample_std, 1000)
    y1 = stats.norm.pdf(x1, loc=population_mean, scale=sample_std / np.sqrt(sample_size))

    # Type 2 Error (Beta): False negative (failing to reject the false null hypothesis)
    z_beta = (population_mean - sample_mean) / (sample_std / np.sqrt(sample_size))
    beta = stats.norm.cdf(z_critical - z_beta)

    # Plot the distribution
    plt.plot(x1, y1, label="Population Distribution", color="blue")
    plt.fill_between(x1, y1, where=(x1 <= population_mean - z_critical * sample_std / np.sqrt(sample_size)) |
                                         (x1 >= population_mean + z_critical * sample_std / np.sqrt(sample_size)), color="red", alpha=0.3, label="Type 1 Error (α)")

    plt.axvline(sample_mean, color="green", linestyle="--", label=f"Sample Mean = {sample_mean}")
    plt.title("Type 1 and Type 2 Errors Visualization")
    plt.xlabel("Value")
    plt.ylabel("Probability Density")
    plt.legend()
    plt.grid(True)
    plt.show()

    print(f"Type 1 Error (Alpha): Rejecting the true null hypothesis with a probability of {alpha}")
    print(f"Type 2 Error (Beta): Failing to reject the false null hypothesis with a probability of {beta:.4f}")

# Example usage
visualize_type_errors(sample_mean=52, population_mean=50, sample_size=30)
```

```python
# 6 Perform an Independent T-test and Interpret the Results
import numpy as np
from scipy import stats

# Simulate two independent datasets
np.random.seed(42)
data1 = np.random.normal(loc=50, scale=10, size=100)  # Group 1
data2 = np.random.normal(loc=55, scale=15, size=100)  # Group 2

# Perform independent T-test
t_stat, p_value = stats.ttest_ind(data1, data2)

print(f"T-statistic: {t_stat:.2f}")
print(f"P-value: {p_value:.4f}")

# Interpretation
alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis: The two groups have significantly different means.")
else:
    print("Fail to reject the null hypothesis: No significant difference between the groups.")
```

```python
# 7. Perform a Paired Sample T-test and Visualize the Comparison Results
import matplotlib.pyplot as plt
import numpy as np
from scipy import stats

# Simulate paired data (before and after treatment)
np.random.seed(42)
before = np.random.normal(loc=50, scale=10, size=30)  # Before treatment
after = before + np.random.normal(loc=5, scale=5, size=30)  # After treatment (added improvement)

# Perform paired sample T-test
t_stat, p_value = stats.ttest_rel(before, after)

print(f"T-statistic: {t_stat:.2f}")
print(f"P-value: {p_value:.4f}")

# Plotting the before and after comparison
plt.figure(figsize=(8, 5))
plt.plot(range(len(before)), before, label="Before Treatment", marker='o', linestyle='--')
plt.plot(range(len(after)), after, label="After Treatment", marker='o', linestyle='-')
plt.title("Paired Sample T-test: Before vs After Treatment")
plt.xlabel("Index")
plt.ylabel("Value")
plt.legend()
plt.grid(True)
plt.show()

# Interpretation
alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis: There is a significant difference between before and after treatment.")
else:
    print("Fail to reject the null hypothesis: No significant difference between before and after treatment.")
```

```python
# 8 Simulate Data and Perform Both Z-test and T-test, Then Compare the Results
import numpy as np
from scipy import stats

# Simulate sample data
np.random.seed(42)
data = np.random.normal(loc=50, scale=10, size=30)  # Sample of 30 observations
population_mean = 52

# Z-test
sample_mean = np.mean(data)
sample_std = np.std(data, ddof=1)
sample_size = len(data)
z_score = (sample_mean - population_mean) / (sample_std / np.sqrt(sample_size))
z_p_value = 2 * (1 - stats.norm.cdf(abs(z_score)))

# T-test
t_stat, t_p_value = stats.ttest_1samp(data, population_mean)

# Output results
print(f"Z-test results: Z-score = {z_score:.2f}, P-value = {z_p_value:.4f}")
print(f"T-test results: T-statistic = {t_stat:.2f}, P-value = {t_p_value:.4f}")

# Interpretation
alpha = 0.05
if z_p_value < alpha:
    print("Reject the null hypothesis in Z-test: The sample mean is significantly different from the population mean.")
else:
    print("Fail to reject the null hypothesis in Z-test: No significant difference.")

if t_p_value < alpha:
    print("Reject the null hypothesis in T-test: The sample mean is significantly different from the population mean.")
else:
    print("Fail to reject the null hypothesis in T-test: No significant difference.")
```

```python
# 9 Write a Python Function to Calculate the Confidence Interval for a Sample Mean and Explain Its Significance
import numpy as np
from scipy import stats

def confidence_interval(data, confidence=0.95):
    sample_mean = np.mean(data)
    sample_size = len(data)
    sample_std = np.std(data, ddof=1)

    # Calculate the margin of error using the t-distribution
    alpha = 1 - confidence
    t_critical = stats.t.ppf(1 - alpha/2, df=sample_size-1)
    margin_of_error = t_critical * (sample_std / np.sqrt(sample_size))

    # Confidence Interval
    lower_bound = sample_mean - margin_of_error
    upper_bound = sample_mean + margin_of_error

    return (lower_bound, upper_bound)

# Example usage
data = np.random.normal(loc=50, scale=10, size=100)
ci = confidence_interval(data)

print(f"95% Confidence Interval: {ci}")
```

```python
# 10. Write a Python Program to Calculate the Margin of Error for a Given Confidence Level Using Sample Data
import numpy as np
from scipy import stats

def margin_of_error(data, confidence=0.95):
    sample_mean = np.mean(data)
    sample_size = len(data)
    sample_std = np.std(data, ddof=1)

    # Critical value for the t-distribution
    alpha = 1 - confidence
    t_critical = stats.t.ppf(1 - alpha/2, df=sample_size-1)

    # Margin of error calculation
    margin_error = t_critical * (sample_std / np.sqrt(sample_size))

    return margin_error

# Example usage
data = np.random.normal(loc=50, scale=10, size=100)
moe = margin_of_error(data)

print(f"Margin of Error at 95% confidence level: {moe:.2f}")
```

```python
# 12. Perform a Chi-square Test for Independence Between Two Categorical Variables in Python
import numpy as np
from scipy.stats import chi2_contingency

# Example data: A 2x2 contingency table
data = np.array([[30, 10], [5, 55]])

# Perform Chi-square test for independence
chi2_stat, p_value, dof, expected = chi2_contingency(data)

# Output results
print(f"Chi-square statistic: {chi2_stat:.2f}")
print(f"P-value: {p_value:.4f}")
print(f"Degrees of freedom: {dof}")
print(f"Expected frequencies: \n{expected}")

# Interpretation
alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis: The variables are dependent.")
else:
    print("Fail to reject the null hypothesis: The variables are independent.")
```

```python
# 13 Write a Python Program to Calculate the Expected Frequencies for a Chi-square Test Based on Observed Data
import numpy as np

# Example observed data (contingency table)
observed = np.array([[30, 10], [5, 55]])

# Calculate row and column totals
row_totals = observed.sum(axis=1, keepdims=True)
col_totals = observed.sum(axis=0, keepdims=True)

# Calculate grand total (sum of all observations)
grand_total = observed.sum()

# Calculate expected frequencies
expected = (row_totals @ col_totals) / grand_total

# Output expected frequencies
print("Expected frequencies:\n", expected)
```

```python
# 14 Perform a Goodness-of-Fit Test Using Python to Compare the Observed Data to an Expected Distribution
import numpy as np
from scipy.stats import chisquare

# Example observed data (counts of categories)
observed = np.array([50, 30, 20])

# Example expected data (expected counts for each category)
expected = np.array([40, 40, 20])

# Perform Goodness-of-Fit test (Chi-square test)
chi2_stat, p_value = chisquare(observed, expected)

print(f"Chi-square statistic: {chi2_stat:.2f}")
print(f"P-value: {p_value:.4f}")

# Interpretation
alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis: The observed data does not fit the expected distribution.")
else:
    print("Fail to reject the null hypothesis: The observed data fits the expected distribution.")
```
